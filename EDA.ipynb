```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
#from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pickle
```

```python
df = pd.read_csv('./dataset/BostonHousing.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df["rm"].fillna(df["rm"].mean(), inplace=True)
```

```python
df.corr()
```

```python
plt.scatter(df['age'], df['medv'])
plt.xlabel("Age")
plt.ylabel("Price")
```

```python
X = df.drop('medv', axis=1)
Y = df['medv']
```

```python
print("Missing values in X is ", X.isnull().sum())
print("Missig values in Y is ", Y.isnull().sum())
```

```python
#imputer = SimpleImputer(strategy='mean')
#X_imputed = imputer.fit_transform(X)
```

```python
X_train, X_test, Y_train, Y_test =  train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train, Y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
lr.get_params()
```

```python
reg_pred = lr.predict(X_test)
```

```python
plt.scatter(reg_pred,Y_test)
```

```python
residuals = Y_test-reg_pred
```

```python
residuals
```

```python
sns.displot(residuals,kind='kde')
```

```python
print(mean_squared_error(Y_test, reg_pred))
print(mean_absolute_error(Y_test, reg_pred))
print(np.sqrt(mean_squared_error(Y_test,reg_pred)))
```

```python
score = r2_score(Y_test, reg_pred)
print(score)
```

```python
pickle.dump(lr, open('reg_model.pkl', 'wb'))
```
